불용어 처리

```python
def erase(x): # 이 함수에다가 삭제할 불용어 추가할 것
  x = str(x)
  x = x.replace("이 책은", "")
  x = x.replace("nan", "")
  return x
```

```python
data['soup'] = data['soup'].apply(erase)
data.head()
```

```python
def make_keyword(x):
  x = str(x)
  li1 = x.split(" ")
  li3 = []
  for a in li1:
    li3.append(mecab.pos(a))
  
  final = []

  for a in li3:
    word = ''
    length = 0
    for b in a:
      if b[1] == 'NNP' or b[1] == 'NNG':
        final.append(b[0])
        word = word + b[0]
        length = len(b[0])
    if len(word) > length:
      final.append(word)
  
  set_final = set(final)
  final = list(set_final)

  return final
```
